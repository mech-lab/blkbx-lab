# 00_qwen35_attach
Generate a canonical `Qwen3.5` reference trace and emit JSON artifacts.

```python
from pathlib import Path

from hybrid_mechlab import HybridLab, profiles
from hybrid_mechlab.topology.offline import compute_persistence, export_trace_and_persistence_artifacts

artifacts_dir = Path('artifacts') / 'qwen35_attach'
lab = HybridLab.attach(model='dummy-qwen', profile=profiles.reference.qwen35(), backend='adapter')
trace = lab.run(prompts=['Track the entity across a long context.'], capture=['codes', 'sketches', 'transport'])
report = compute_persistence(trace)
paths = export_trace_and_persistence_artifacts(trace, str(artifacts_dir))
print(trace.summary())
print(report.summary)
print(paths)
```
