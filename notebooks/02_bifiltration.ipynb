# 02_bifiltration
Attach interventions, run a trace, and inspect the resulting persistence summary.

```python
from hybrid_mechlab import HybridLab, profiles
from hybrid_mechlab.hooks.interventions import FeatureClamp, BridgeGateScale
from hybrid_mechlab.topology.offline import compute_persistence

interventions = [
    FeatureClamp(target='block.1', feature_id=7, value=0.0),
    BridgeGateScale(target='block.bridge', scale=0.5),
]
trace = HybridLab.attach(model='dummy-qwen', profile=profiles.reference.qwen35(), backend='adapter').run(
    prompts=['Evaluate intervention sensitivity over bridge cadence.'],
    interventions=interventions,
)
report = compute_persistence(trace)
print(trace.summary())
print(report.summary)
```
