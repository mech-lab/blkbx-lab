# 04_ratio_comparison
Export the canonical `Qwen3.5` vs `Gated DeltaNet` persistence comparison JSON.

```python
from pathlib import Path

from hybrid_mechlab import HybridLab, profiles
from hybrid_mechlab.topology.offline import compute_persistence, compare_persistence, export_persistence_comparison

artifacts_dir = Path('artifacts') / 'qwen_vs_gated'
qwen_trace = HybridLab.attach(model='dummy-qwen', profile=profiles.reference.qwen35(), backend='adapter').run(
    prompts=['Compare a reference profile against a native kernel.']
)
gated_trace = HybridLab.attach(model='dummy-native', profile=profiles.native.gated_deltanet(), backend='native').run(
    prompts=['Compare a native kernel against a reference profile.']
)
qwen_report = compute_persistence(qwen_trace)
gated_report = compute_persistence(gated_trace)
comparison = compare_persistence(qwen_report, gated_report)
artifact_path = artifacts_dir / 'comparison.json'
export_persistence_comparison(comparison, str(artifact_path))
print(comparison)
print(artifact_path)
```
